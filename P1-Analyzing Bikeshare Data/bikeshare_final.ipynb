```python
## TODO: import all necessary packages and functions
#import pandas as pd
#import numpy as np
#import datetime as dt
import pandas as pd
import datetime as dt
import numpy as np
import time

## Filenames
chicago = 'chicago.csv'
new_york_city = 'new_york_city.csv'
washington = 'washington.csv'

pd.options.mode.chained_assignment = None

def get_city():
    '''Asks the user for a city and returns the filename for that city's bike share data.

    Args:
        none.
    Returns:
        (str) Filename for a city's bikeshare data.
    '''
    city = input('\nHello! Let\'s explore some US bikeshare data!\n'
                 'Would you like to see data for Chicago, New York, or Washington?\n')
    # TODO: handle raw input and complete function
    if city.lower() == 'chicago':
        city_file = pd.read_csv(chicago)
    elif city.lower() == "new york":
        city_file = pd.read_csv(new_york_city)
    elif city.lower() == 'washington':
        city_file = pd.read_csv(washington)
    else:
        print('\nPlease enter a valid city name\n')
        get_city()
    return city_file
   # return city

def get_time_period():
    '''Asks the user for a time period and returns the specified filter.

    Args:
        none.
    Returns: time period
        TODO: fill out return type and description (see get_city for an example)
    '''
    time_period = input('\nWould you like to filter the data by month, day, or not at'
                        ' all? Type "none" for no time filter.\n')
    # TODO: handle raw input and complete function\
    if time_period.lower() == 'month':
        return 'month'
    elif time_period.lower() == 'day':
        return 'day'
    elif time_period == 'none':
        return 'none'
    else:
        print('Enter a response from the selection')
        return get_time_period()

def get_month():
    '''Asks the user for a month and returns the specified month.

    Args:
        none.
    Returns: specified month
        TODO: fill out return type and description (see get_city for an example)
    '''
    month = input('\nWhich month? January, February, March, April, May, or June?\n')
    # TODO: handle raw input and complete function
    if month.lower() == 'january':
        month = 'January'
    elif month.lower() == 'february':
        month = 'February'
    elif month.lower() == 'march':
        month = 'March'
    elif month.lower() == 'april':
        month = 'April'
    elif month.lower() == 'may':
        month = 'May'
    elif month.lower() == 'june':
        month = 'June'
    else:
        print('Sorry please enter valid input')
    return month

def get_day():
    '''Asks the user for a day and returns the specified day.

    Args:
        none.
    Returns: Select day
        TODO: fill out return type and description (see get_city for an example)
    '''
    day = input('\nWhich day? Please type your response as integer. 0 = Monday and 6 = Sunday.\n')
    if day == '0':
        daychoice = 'Monday'
    elif day == '1':
        daychoice = 'Tuesday'
    elif day == '2':
        daychoice = 'Wednesday'
    elif day == '3':
        daychoice = 'Thursday'
    elif day == '4':
        daychoice = 'Friday'
    elif day == '5':
        daychoice = 'Saturday'
    elif day == '6':
        daychoice = 'Sunday'
    else:
        print('Please enter valid response')
        return get_day()
    # TODO: handle raw input and complete function
    print("You've selected "+ daychoice)
    return daychoice

def popular_month(city_file):
    '''Returns most popular month for start time.
    TODO: fill out docstring with description, arguments, and return values.
    Args:
        city_file
    Returns:
       popular_month
    Question: What is the most popular month for start time?
    '''

    #    df = city_data[city_data['month_num'] ==
    #convert to datetime format
    ##city_file['Start Time'] = pd.to_datetime(city_file['Start Time'])
    #df.loc[df['Start Time'].dt.month == month_num]
    #df['MONTH'] = df['Start Time'].dt.month
    #city_file['month'] = city_file['Start Time'].dt.strftime('%B')
    city_file['month'] = city_file['Start Time'].dt.strftime('%B')
    month_mode = city_file['month'].mode()
    return ("\nThe most popular month is "+ month_mode[0])

def popular_day(city_file):
    '''
    Args: city file
    Return: popular day
    TODO: fill out docstring with description, arguments, and return values.
    Question: What is the most popular day of week (Monday, Tuesday, etc.) for start time?
    '''
    # TODO: complete function
    #df = pd.read_csv(city_file)
        #convert to datetime format
    #city_file['Start Time'] = pd.to_datetime(city_file['Start Time'])
    city_file['day'] = city_file['Start Time'].dt.strftime('%A')
    #city_file['DAY'] = city_file['Start Time'].dt.weekday_name
    day_mode = city_file['day'].mode()
    return ("\nThe most popular day is "+ day_mode[0])

def popular_hour(city_file):
    '''
    Args: city file
    Return: popular hour

    Question: What is the most popular hour of day for start time?
    '''
    # TODO: complete function
    #df = pd.read_csv(city_file)
    #city_file['Start Time'] = pd.to_datetime(city_file['Start Time'])
    city_file['hour'] = city_file['Start Time'].dt.strftime('%H')
    hour_mode = city_file['hour'].mode()
    return("\nThe most popular hour is " + hour_mode[0])

def trip_duration(city_file):
    '''Caclculate total and average trip Duration
    Arg: city_data
    Return: total trip duration and average trip duration

    Question: What is the total trip duration and average trip duration?
    '''
    # TODO: complete function
    #df = pd.read_csv(city_file)
    #query for particular df['Trip Duration']
    ttl_trip_oa = city_file['Trip Duration'].sum()
    avg_trip_oa = city_file['Trip Duration'].mean()
    return ("\nThe total trip duration is " + str(round(ttl_trip_oa)) + " and the average trip time is "
                 + str(round(avg_trip_oa)))

def popular_stations(city_file):
    '''Find most popular start and end stations
    Arg: city_data
    Return: Popular start and end stations

    Question: What is the most popular start station and most popular end station?
    '''
    # TODO: complete function
    #df = pd.read_csv(city_file)
#     #convert to datetime format
#     df['Start Time'] = pd.to_datetime(df['Start Time'])
#     df['day'] = df['Start Time'].dt.strftime('%A')
#     #city_file['DAY'] = city_file['Start Time'].dt.weekday_name
    start_mode = city_file['Start Station'].mode()
    end_mode = city_file['End Station'].mode()
    return ('\nThe most popular start station is ' +str(start_mode[0])+ '.\nThe most popular end station is ' +
                 str(end_mode[0]))

def popular_trip(city_file):
    '''Find popular trip
    Arg: city file
    Return: popular trip

    Question: What is the most popular trip?
    '''
    # TODO: complete function
    #df = pd.read_csv(city_file)
    #city_mode= df['Cities'].mode()
    city_file['trip'] = city_file['Start Station'] + ' ====>' + city_file['End Station']
    p_trip = city_file['trip'].mode()
    #print(city_file['trip'])
    return ("\nThe most popular trip is " + p_trip[0])

def users(city_file):
    '''Display counts of a each user type
    Arg: city_file
    Return: User counts
    Question: What are the counts of each user type?
    '''
    # TODO: complete function
    #df = pd.read_csv(city_file)
    user_types = city_file['User Type'].value_counts().to_frame()
    #print("Counts of each User type")
    return(user_types)

def gender(city_file):
    '''Display counts of Gender
        Arg: city_file
        Return: gender counts
    Question: What are the counts of gender?
    '''
    # TODO: complete function
    #df1 = pd.read_csv(city_file)
    gender = city_file['Gender'].value_counts().to_frame()
    #print ("\nGender Counts")
    return(gender)

def birth_years(city_file):
    '''Find earliest, most recent, and most popular birth years.
        Arg: city_file
        Return: Min, max, mode of birth year based on fitlered data

    Question: What are the earliest (i.e. oldest user), most recent (i.e. youngest user),
    and most popular birth years?
    '''
    # TODO: complete function
    #df = pd.read_csv(city_file)
    by_max = city_file['Birth Year'].max()
    by_min = city_file['Birth Year'].min()
    by_mode = city_file['Birth Year'].mode()
    return( '\nThe most recent year is ' + str(by_max)+'.\nThe oldest is ' + str(by_min)+' .\nThe most popular birth year is '+ str(round(by_mode[0])))

def display_data(city_file):
    '''Displays five lines of data if the user specifies that they would like to.
    After displaying five lines, ask the user if they would like to see five more,
    continuing asking until they say stop.

    Args: City file data
         none.
    Returns: Filtered data
    '''
    display = input('\nWould you like to view individual trip data?' 'Type \'yes\' or \'no\'.\n')
    if display == 'yes':
        s = 0
        f = 5
        print(city_file.iloc[s:f])
        while True:
            answer = input('\nWould you like to view the next 5 lines of data?' 'Type \'yes\' or \'no\'.\n')
            if answer=="yes":
                s += 5
                f += 5
                print(city_file.iloc[s:f])
                continue
            elif answer=="no":
                break
    elif display == 'no':
        exit
    else:
        print('Please enter valid response')
        display_data(city_file)

#Dictionar yof different data filters and corresponding functions
filter_options = {'none':  [popular_month, popular_day, popular_hour, trip_duration, popular_stations, popular_trip, users, gender, birth_years],
'month_filter': [popular_day, popular_hour, trip_duration, popular_stations, popular_trip, users, gender, birth_years],
'day_filter': [popular_hour, trip_duration, popular_stations, popular_trip, users, gender, birth_years,]}

def statistics():
    '''Calculates and prints out the descriptive statistics about a city and time period
    specified by the user via raw input.

    Args:
        none.
    Returns:
        Calculated statistics basesd on select filter
    '''
    # Filter by city (Chicago, New York, Washington)
    city_file = get_city()
    #Convert Start and End Time columns to datetime format
    city_file['Start Time'] = pd.to_datetime(city_file['Start Time'])
    city_file['End Time'] = pd.to_datetime(city_file['End Time'])
    # Filter by time period (month, day, none)
    time_period = get_time_period()

    if time_period == 'month':
        month = get_month()
        #city_file['Start Time'] = pd.to_datetime(city_file['Start Time'])
        filter_month_data = city_file[city_file['Start Time'].dt.month==dt.datetime.strptime(month, '%B').month]
        for i in filter_options['month_filter']:
            start_time = time.time()
            print('\nCalculating  statistic...')
            print(i(filter_month_data))
            print("\nThat took %s seconds." % (time.time() - start_time))

    elif time_period == 'day':
        day = get_day()
        #city_file['Start Time'] = pd.to_datetime(city_file['Start Time'])
        filter_day_data = city_file[city_file['Start Time'].dt.day==dt.datetime.strptime(day, '%A').day]
        for i in filter_options['day_filter']:
            start_time = time.time()
            print('\nCalculating statistic...')
            print(i(filter_day_data))
            print("\nThat took %s seconds." % (time.time() - start_time))
            print('\n')
    else:
        for i in filter_options['none']:
            #city_file['Start Time'] = pd.to_datetime(city_file['Start Time'])
            start_time = time.time()
            print('\nCalculating statistic...')
            print(i(city_file))
            print('\nThat took %s seconds.' % (time.time() - start_time))
            print('\n')

    #Ask user if they want to display data
    display_data(city_file)

    # Restart?
    restart = input('\nWould you like to restart? Type \'yes\' or \'no\'.\n')
    if restart.lower() == 'yes':
        statistics()


if __name__ == "__main__":
	statistics()
```
